```python
hyperparameters = {
    "learning_rate": 1e-4,
    "batch_size": 64,
    "embed_dim": 128,
    "num_heads": 4,
    "num_layers": 2,
    "dropout": 0.1,
    "ff_dim": 1024,
    "epochs": 300,
    "weight_decay": 1e-4,
    "num_warmup_steps": 500,
    "k-mers": 3,
    "max_len": 199,
}

info = {
    "dataset_size": "50K",
    "precision": "FP16",
    "dir_name": "Ref_Alt_MutType_Model",
    "run": "First Run"
}
```

```python
import pandas as pd

data_path = f"/content/drive/MyDrive/dataset/{info['dataset_size']}.csv"

data = pd.read_csv(data_path)
data.shape
```

```python
data['ref'] = data['ref'].map({'A': 0, 'T': 1, 'G': 2, 'C': 3})
data['alt'] = data['alt'].map({'A': 0, 'T': 1, 'G': 2, 'C': 3})
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['mutation_type'] = encoder.fit_transform(data['mutation_type'])
```

```python
x = data['sequence']
y_ref = data['ref']
y_alt = data['alt']
y_type = data['mutation_type']
```

```python
def get_codon(seq, k=hyperparameters['k-mers']):
    return [seq[i:i+k] for i in range(len(seq) - k + 1)]

vocab = {'<PAD>': 0, '<UNK>': 1}

for seq in data['sequence']:
    for codons in get_codon(seq.lower()):
        if codons not in vocab:
            vocab[codons] = len(vocab)
        else:
            continue

def get_tensor(text):
    return [vocab.get(codons.lower(), vocab['<UNK>']) for codons in get_codon(text)]
```

```python
import torch
from torch.utils.data import Dataset

class CustomDataset(Dataset):
  def __init__(
      self,
      x,
      y_ref,
      y_alt,
      y_type
  ):
    self.x_frame = x
    self.y_ref_frame = y_ref
    self.y_alt_frame = y_alt
    self.y_type_frame = y_type

  def __len__(self):
    return len(self.x_frame)

  def __getitem__(self, index):
    x = torch.tensor(get_tensor(self.x_frame[index]), dtype=torch.long)
    y_ref = torch.tensor(self.y_ref_frame[index], dtype=torch.long)
    y_alt = torch.tensor(self.y_alt_frame[index], dtype=torch.long)
    y_type = torch.tensor(self.y_type_frame[index], dtype=torch.long)

    return (
      x,
      y_ref,
      y_alt,
      y_type
    )
```

```python
dataset = CustomDataset(
    x,
    y_ref,
    y_alt,
    y_type
)
```

```python
from torch.utils.data import random_split

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size

train_dataset, test_dataset = random_split(dataset, [train_size, test_size])
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(
    train_dataset,
    batch_size=hyperparameters['batch_size'],
    shuffle=True
)
test_loader = DataLoader(
    test_dataset,
    batch_size=hyperparameters['batch_size']
)
```

```python
from torch import nn
import math

class PositionalEncoding(nn.Module):
    def __init__(self, embed_dim, max_len=5000):
        super().__init__()

        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len).unsqueeze(1)

        div_term = torch.exp((torch.arange(0, embed_dim, 2)) * (-math.log(10000.0) / embed_dim))

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)

        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        x = x + self.pe[:, :x.size(1), :].to(x.device)
        return x

class Transformer(nn.Module):
    def __init__(self, embed_dim=512, num_heads=8, num_layers=6, ff_dim=2048, dropout=0.1, vocab_size=10000, max_len=5000):
        super(Transformer, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embed_dim)
        self.position_encoding = PositionalEncoding(embed_dim=embed_dim, max_len=max_len)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=num_heads,
            dim_feedforward=ff_dim,
            dropout=dropout,
            batch_first=True
        )
        self.encoder = nn.TransformerEncoder(
            encoder_layer=encoder_layer,
            num_layers=num_layers
        )

        self.y_ref_out = nn.Linear(embed_dim, 4)
        self.y_alt_out = nn.Linear(embed_dim, 4)
        self.y_mut_out = nn.Linear(embed_dim, 12)

    def forward(self, x):
        x = self.embeddings(x)
        x = self.position_encoding(x)

        x = self.encoder(x)
        x = x.mean(dim=1)

        y_ref_out = self.y_ref_out(x)
        y_alt_out = self.y_alt_out(x)
        y_mut_out = self.y_mut_out(x)

        return (
            y_ref_out,
            y_alt_out,
            y_mut_out
        )
```

```python
model = Transformer(
    embed_dim=hyperparameters['embed_dim'],
    num_heads=hyperparameters['num_heads'],
    num_layers=hyperparameters['num_layers'],
    ff_dim=hyperparameters['ff_dim'],
    dropout=hyperparameters['dropout'],
    vocab_size=len(vocab),
    max_len=hyperparameters['max_len']
)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total trainable parameters: {num_params}")
```

```python
from torch import optim
from transformers import get_linear_schedule_with_warmup

ce_ref = nn.CrossEntropyLoss()
ce_alt = nn.CrossEntropyLoss()
ce_mut = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(), lr=hyperparameters['learning_rate'], weight_decay=hyperparameters['weight_decay'])
scaler = torch.cuda.amp.GradScaler()

num_training_steps = len(train_loader) * hyperparameters['epochs']
num_warmup_steps = hyperparameters['num_warmup_steps']

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=num_training_steps
)
```

```python
def train32(
    model,
    loader,
    ce_ref,
    ce_alt,
    ce_mut,
    optimizer
):

    model.train()

    running_loss, correct_ref, correct_alt, correct_mut, total = 0.0, 0, 0, 0, 0

    for x, y_ref, y_alt, y_mut in loader:
        optimizer.zero_grad()

        x = x.to(device)
        y_ref = y_ref.to(device)
        y_alt = y_alt.to(device)
        y_mut = y_mut.to(device)

        y_ref_out, y_alt_out, y_mut_out = model(x)

        loss_ref = ce_ref(y_ref_out, y_ref)
        loss_alt = ce_alt(y_alt_out, y_alt)
        loss_mut = ce_mut(y_mut_out, y_mut)

        loss = loss_ref + loss_alt + loss_mut

        pred_ref = torch.argmax(y_ref_out, dim=1)
        correct_ref += (pred_ref == y_ref).sum().item()

        pred_alt = torch.argmax(y_alt_out, dim=1)
        correct_alt += (pred_alt == y_alt).sum().item()

        pred_mut = torch.argmax(y_mut_out, dim=1)
        correct_mut += (pred_mut == y_mut).sum().item()

        total += len(x)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * len(x)

    accuracy_ref = correct_ref / total
    accuracy_alt = correct_alt / total
    accuracy_mut = correct_mut / total

    return (
        running_loss/len(loader.dataset),
        accuracy_ref,
        accuracy_alt,
        accuracy_mut,
    )

def train16(
        model,
        loader,
        ce_ref,
        ce_alt,
        ce_mut,
        optimizer,
        scaler,
        scheduler
    ):

    model.train()

    running_loss, correct_ref, correct_alt, correct_mut, total = 0.0, 0, 0, 0, 0

    for x, y_ref, y_alt, y_mut in loader:
        optimizer.zero_grad()

        x = x.to(device)
        y_ref = y_ref.to(device)
        y_alt = y_alt.to(device)
        y_mut = y_mut.to(device)

        with torch.cuda.amp.autocast():
            y_ref_out, y_alt_out, y_mut_out = model(x)

            loss_ref = ce_ref(y_ref_out, y_ref)
            loss_alt = ce_alt(y_alt_out, y_alt)
            loss_mut = ce_mut(y_mut_out, y_mut)

            loss = loss_ref + loss_alt + loss_mut

        pred_ref = torch.argmax(y_ref_out, dim=1)
        correct_ref += (pred_ref == y_ref).sum().item()

        pred_alt = torch.argmax(y_alt_out, dim=1)
        correct_alt += (pred_alt == y_alt).sum().item()

        pred_mut = torch.argmax(y_mut_out, dim=1)
        correct_mut += (pred_mut == y_mut).sum().item()

        total += len(x)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        scheduler.step()

        running_loss += loss.item() * len(x)


    accuracy_ref = correct_ref / total
    accuracy_alt = correct_alt / total
    accuracy_mut = correct_mut / total

    return (
        running_loss/len(loader.dataset),
        accuracy_ref,
        accuracy_alt,
        accuracy_mut,
    )
```

```python
def validation(
        model,
        loader,
        ce_ref,
        ce_alt,
        ce_mut
):
    model.eval()

    running_loss, correct_ref, correct_alt, correct_mut, total = 0.0, 0, 0, 0, 0

    with torch.no_grad():
        for x, y_ref, y_alt, y_mut in loader:
                x = x.to(device)
                y_ref = y_ref.to(device)
                y_alt = y_alt.to(device)
                y_mut = y_mut.to(device)

                if info['precision'] == 'FP16':
                        with torch.cuda.amp.autocast():
                                y_ref_out, y_alt_out, y_mut_out = model(x)

                                loss_ref = ce_ref(y_ref_out, y_ref)
                                loss_alt = ce_alt(y_alt_out, y_alt)
                                loss_mut = ce_mut(y_mut_out, y_mut)

                                loss = loss_ref + loss_alt + loss_mut
                else:
                        y_ref_out, y_alt_out, y_mut_out = model(x)

                        loss_ref = ce_ref(y_ref_out, y_ref)
                        loss_alt = ce_alt(y_alt_out, y_alt)
                        loss_mut = ce_mut(y_mut_out, y_mut)

                        loss = loss_ref + loss_alt + loss_mut

                running_loss += loss.item() * len(x)

                pred_ref = torch.argmax(y_ref_out, dim=1)
                correct_ref += (pred_ref == y_ref).sum().item()

                pred_alt = torch.argmax(y_alt_out, dim=1)
                correct_alt += (pred_alt == y_alt).sum().item()

                pred_mut = torch.argmax(y_mut_out, dim=1)
                correct_mut += (pred_mut == y_mut).sum().item()

                total += len(x)

        accuracy_ref = correct_ref / total
        accuracy_alt = correct_alt / total
        accuracy_mut = correct_mut / total

        return (
                running_loss/len(loader.dataset),
                accuracy_ref,
                accuracy_alt,
                accuracy_mut,
        )
```

```python
import os

patience = 10
best_val_loss = float('inf')
counter = 0
early_stop = False

train_loss_history = []
train_ref_acc_history = []
train_alt_acc_history = []
train_mut_acc_history = []

validation_loss_history = []
validation_ref_acc_history = []
validation_alt_acc_history = []
validation_mut_acc_history = []

save_dir = f"/content/drive/MyDrive/{info['dir_name']}"
os.makedirs(save_dir, exist_ok=True)

for epoch in range(1, hyperparameters['epochs']+1):
    if info['precision'] == 'FP16':
      train_loss, train_accuracy_ref, train_accuracy_alt, train_accuracy_mut = train16(
            model,
            train_loader,
            ce_ref,
            ce_alt,
            ce_mut,
            optimizer,
            scaler,
            scheduler
        )
    else:
      train_loss, train_accuracy_ref, train_accuracy_alt, train_accuracy_mut = train32(
          model,
          train_loader,
          ce_ref,
          ce_alt,
          ce_mut,
          optimizer
      )

    validation_loss, validation_accuracy_ref, validation_accuracy_alt, validation_accuracy_mut = validation(
        model,
        test_loader,
        ce_ref,
        ce_alt,
        ce_mut
    )

    train_loss_history.append(train_loss)
    train_ref_acc_history.append(train_accuracy_ref)
    train_alt_acc_history.append(train_accuracy_alt)
    train_mut_acc_history.append(train_accuracy_mut)

    validation_loss_history.append(validation_loss)
    validation_ref_acc_history.append(validation_accuracy_ref)
    validation_alt_acc_history.append(validation_accuracy_alt)
    validation_mut_acc_history.append(validation_accuracy_mut)

    print(f"""
    Epoch ({epoch}/{hyperparameters['epochs']}):
    ----------------------------------------
    Training:
    Loss                      = {train_loss:.4f}
    Accuracy (Reference)      = {train_accuracy_ref:.2f}%
    Accuracy (Alternate)      = {train_accuracy_alt:.2f}%
    Accuracy (Mutation Type)  = {train_accuracy_mut:.2f}%

    Validation:
    Loss                      = {validation_loss:.4f}
    Accuracy (Reference)      = {validation_accuracy_ref:.2f}%
    Accuracy (Alternate)      = {validation_accuracy_alt:.2f}%
    Accuracy (Mutation Type)  = {validation_accuracy_mut:.2f}%
    """)

    if (epoch+1) % 10 == 0:
      checkpoint_path = f"{save_dir}/model_{info['run']}_epoch_{epoch+1}.pth"
      torch.save({
          'epoch': epoch+1,
          'model_state_dict': model.state_dict(),
          'optimizer_state_dict': optimizer.state_dict(),
          'Training Loss': train_loss,
          'Training Accuracy Reference': train_accuracy_ref,
          'Training Accuracy Alternate': train_accuracy_alt,
          'Training Accuracy Muation': train_accuracy_mut,
          'Validation Loss': validation_loss,
          'Validation Accuracy Reference': validation_accuracy_ref,
          'Validation Accuracy Alternate': validation_accuracy_alt,
          'Validation Accuracy Muation': validation_accuracy_mut
      }, checkpoint_path)
      print(f"Model saved at {checkpoint_path}")


    if validation_loss < best_val_loss:
        best_val_loss = validation_loss
        counter = 0
        continue
    else:
        counter += 1
        print(f"No improvement in val loss Counter = {counter}/{patience}")
        if counter >= patience:
            print("Early stopping triggered!")
            early_stop = True
            break
```

```python
from sklearn.metrics import classification_report

def get_predictions_and_labels(model, loader):
    model.eval()
    all_y_true_ref = []
    all_y_true_alt = []
    all_y_true_mut = []

    all_y_pred_ref = []
    all_y_pred_alt = []
    all_y_pred_mut = []

    with torch.no_grad():
        for x, y_ref, y_alt, y_mut in loader:
            x = x.to(device)
            y_ref = y_ref.to(device)
            y_alt= y_alt.to(device)
            y_mut = y_mut.to(device)

            y_ref_out, y_alt_out, y_mut_out = model(x)

            _, pred_ref = torch.max(y_ref_out, 1)
            _, pred_alt = torch.max(y_alt_out, 1)
            _, pred_mut = torch.max(y_mut_out, 1)

            all_y_true_ref.extend(y_ref.cpu().numpy())
            all_y_true_alt.extend(y_alt.cpu().numpy())
            all_y_true_mut.extend(y_mut.cpu().numpy())

            all_y_pred_ref.extend(pred_ref.cpu().numpy())
            all_y_pred_alt.extend(pred_alt.cpu().numpy())
            all_y_pred_mut.extend(pred_mut.cpu().numpy())

    return (
        (all_y_true_ref, all_y_pred_ref),
        (all_y_true_alt, all_y_pred_alt),
        (all_y_true_mut, all_y_pred_mut)
    )
```

```python
from sklearn.metrics import classification_report

(y_ref_true, y_ref_pred), (y_alt_true, y_alt_pred), (y_mut_true, y_mut_pred) = get_predictions_and_labels(model, test_loader)

print("\n" + "="*60)
print("Classification Report Summary")
print("="*60)

print("\n[1] Classification Report  Reference Label")
print("-" * 60)
print(classification_report(y_ref_true, y_ref_pred))

print("\n[2] Classification Report  Alternate Label")
print("-" * 60)
print(classification_report(y_alt_true, y_alt_pred))

print("\n[3] Classification Report  Mutation Type Label")
print("-" * 60)
print(classification_report(y_mut_true, y_mut_pred))

print("="*60 + "\n")
```
